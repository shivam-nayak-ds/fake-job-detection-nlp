```python
text_columns = [
    "title",
    "location",
    "company_profile",
    "description",
    "requirements",
    "benefits"
]

df[text_columns] = df[text_columns].fillna("")
```

```python
import re
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
## lowercasing
df['text'] = df['text'].str.lower()

## removing html tags
df["text"] = df["text"].apply(lambda x: re.sub(r"<.*?>"," ",x))

## remove special numbers
df["text"] = df["text"].apply(lambda x: re.sub(r"[^a-zA-Z\s]"," ",x))
```

```python
df.loc[2,'text']
```

```python
## tokenization

df['tokens'] = df['text'].apply(word_tokenize)
```

```python
df.head(2)
```

```python
df[["text","tokens"]].head()
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
df['token'] = df['tokens'].apply(
  lambda words: [w for w in words if w not in stop_words]
)
```

```python
from nltk.stem import WordNetLemmatizer
Lematize = WordNetLemmatizer()
df["token"] = df["tokens"].apply(
    lambda words: [Lematize.lemmatize(w) for w in words]
)
```

```python
df.head(2)
```

```python
df.rename(columns={"token": "lemma_tokens"}, inplace=True)
```

```python
df['clean_text'] = df['lemma_tokens'].apply(lambda x: " ".join(x))
```

```python
df[["text","clean_text"]].head()
```

```python
df = df.drop(columns=["tokens","lemma_tokens" , "cleam_text"], errors="ignore")
```

## Text Preprocessing

Text preprocessing is performed to clean and normalize the job posting text so that it can be effectively used by machine learning models.

The following steps were applied:

* **Lowercasing:** Converted all text to lowercase to maintain consistency.
* **Removing special characters:** Removed punctuation and unnecessary symbols.
* **Tokenization:** Split text into individual words (tokens).
* **Stopwords removal:** Removed common words such as *the, is, and* that do not add meaningful information.
* **Lemmatization:** Converted words to their base form (e.g., *running → run*).

Finally, the processed tokens were combined back into a cleaned text column called **`clean_text`**, which will be used for feature extraction in the modeling phase.
